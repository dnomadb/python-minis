```python
import requests, json
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import rasterio
```

```python
reqURL = 'https://data.sfgov.org/api/views/tmnf-yvry/rows.json?accessType=DOWNLOAD'
```

```python
requ = requests.get(reqURL)
reqJSON = requ.json()
```

```python
def latlngToXY(ll):
    from math import pi, log, tan
    D2R = pi / 180
    A = 6378137
    x = (A * ll[0] * D2R)
    y = (A * log(tan((pi*0.25) + (0.5 * ll[1] * D2R))))
    return x,y
```

```python
xys = np.array([latlngToXY((float(i[-1][2]),float(i[-1][1]))) for i in reqJSON['data']])
```

```python
xmin = xys[:,0].min()
xmax = xys[:,0].max()
ymin = xys[:,1].min()
ymax = xys[:,1].max()
X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
```

```python
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([xys[:,0], xys[:,1]])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, X.shape)
```

```python
Z16 = (Z*10e+10).astype(np.uint16)
Z16.max()
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
ax.imshow(np.rot90(Z16), cmap='pink', extent=[xmin, xmax, ymin, ymax])
ax.plot(xys[:,0], xys[:,1], 'r.', markersize=0.5)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
```

## Compute the extent and affine transform matrix for Rasterio output

**NB**: GDAL tracks the upper left corner of pixels, while the grids are centered, so we must adjust slightly.

```python
pixel_size_x = (xmax - xmin)/Z16.shape[0]
pixel_size_y = (ymax - ymin)/Z16.shape[1]
upper_left_x = xmin - pixel_size_x/2.0
upper_left_y = ymax + pixel_size_y/2.0

transform = rasterio.Affine(
                pixel_size_x, 0.0, upper_left_x,
                0.0, -pixel_size_y, upper_left_y)

print(upper_left_x, upper_left_y)
print(pixel_size_x, pixel_size_y)
```

```python
with rasterio.open(
        'crime.tif', 'w', 
        driver='GTiff', 
        dtype=Z16.dtype.type, 
        count=1, 
        width=Z16.shape[1], 
        height=Z16.shape[0], 
        crs={'init': 'EPSG:3857'},
        transform=transform ) as dst:
    # The reformatting as C-contiguous below is a work around for a
    # bug in Rasterio 0.10.
    dst.write_band(1, np.asarray(np.rot90(Z16), order='C'))
```
