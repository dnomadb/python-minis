```python
import requests, json
from scipy import stats
```

```python
reqURL = 'https://data.sfgov.org/api/views/tmnf-yvry/rows.json?accessType=DOWNLOAD'
```

```python
requ = requests.get(reqURL)
reqJSON = requ.json()
```

```python
xys = array([latlngToXY((float(i[-1][2]),float(i[-1][1]))) for i in reqJSON['data']])
```

```python
def latlngToXY(ll):
    from math import pi, log, tan
    D2R = pi / 180
    A = 6378137
    x = (A * ll[0] * D2R)
    y = (A * log(tan((pi*0.25) + (0.5 * ll[1] * D2R))))
    return x,y
```

```python
xmin = xys[:,0].min()
xmax = xys[:,0].max()
ymin = xys[:,1].min()
ymax = xys[:,1].max()
```

```python
X, Y = mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = vstack([X.ravel(), Y.ravel()])
values = vstack([xys[:,0], xys[:,1]])
kernel = stats.gaussian_kde(values)
Z = reshape(kernel(positions).T, X.shape)
```

```python
fig = figure(figsize=(15,15))
ax = fig.add_subplot(111)
ax.imshow(rot90(Z), cmap='rainbow', extent=[xmin, xmax, ymin, ymax])
ax.plot(xys[:,0], xys[:,1], 'r.', markersize=0.5)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
```
