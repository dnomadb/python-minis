```python
import requests, json
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import rasterio
reqURL = 'https://data.sfgov.org/api/views/tmnf-yvry/rows.json?accessType=DOWNLOAD'
requ = requests.get(reqURL)
reqJSON = requ.json()
def latlngToXY(ll):
    from math import pi, log, tan
    D2R = pi / 180
    A = 6378137
    x = (A * ll[0] * D2R)
    y = (A * log(tan((pi*0.25) + (0.5 * ll[1] * D2R))))
    return x,y
def classifyCrime(zArr,classes,weighting=1):
    'convert crime kernel surface into classed surface for polygonization'
    outRas = zeros(zArr.shape)
    zMax = zArr.max()
    zMin = zArr.min()
    zRange = zMax-zMin
    zInterval = zRange/float(classes)
    for i in range(0,classes):
        cClass = ((i*zInterval+zMin)*weighting+percentile(zArr, i/float(classes)*100))/(float(weighting)+1)
        outRas[where(zArr>cClass)] = cClass
    return outRas.astype(np.uint16)
```

```python
xys = np.array([latlngToXY((float(i[-1][2]),float(i[-1][1]))) for i in reqJSON['data']])
xmin = xys[:,0].min()
xmax = xys[:,0].max()
ymin = xys[:,1].min()
ymax = xys[:,1].max()
X, Y = np.mgrid[xmin:xmax:200, ymin:ymax:200]
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([xys[:,0], xys[:,1]])
kernel = stats.gaussian_kde(values,bw_method=0.03)
Z = np.reshape(kernel(positions).T, X.shape)
```

```python
Ztemp = (Z*10e+10).astype(np.uint16)
Z16 = classifyCrime(Ztemp,50,0.5)
#hist(Ztemp.ravel(),bins=50)
# fig = plt.figure(figsize=(15,15))
# ax = fig.add_subplot(111)
# ax.imshow(rot90(Z16), cmap='pink', extent=[xmin, xmax, ymin, ymax])
# ax.plot(xys[:,0], xys[:,1], 'r.', markersize=0.5)
# ax.set_xlim([xmin, xmax])
# ax.set_ylim([ymin, ymax])

hist(Z, bins=50)

# for feature, shapes in features.shapes(asarray(rot90(Z16.astype(uint8)),order='C'),transform=transform):
#     tfeat = []
#     for i in feature['coordinates'][0]:
#         tfeat.append([i[0],i[1]])
#     tfeat = array(tfeat)
#     ax.plot(tfeat[:,0],tfeat[:,1],c='white',lw=0.5)
```

## Compute the extent and affine transform matrix for Rasterio output

**NB**: GDAL tracks the upper left corner of pixels, while the grids are centered, so we must adjust slightly.

```python
pixel_size_x = (xmax - xmin)/Z16.shape[0]
pixel_size_y = (ymax - ymin)/Z16.shape[1]
upper_left_x = xmin - pixel_size_x/2.0
upper_left_y = ymax + pixel_size_y/2.0

transform = rasterio.Affine(
                pixel_size_x, 0.0, upper_left_x,
                0.0, -pixel_size_y, upper_left_y)

print(upper_left_x, upper_left_y)
print(pixel_size_x, pixel_size_y)
```

```python
with rasterio.open(
        'crime.tif', 'w', 
        driver='GTiff', 
        dtype=Z16.dtype.type, 
        count=1, 
        width=Z16.shape[1], 
        height=Z16.shape[0], 
        crs={'init': 'EPSG:3857'},
        transform=transform ) as dst:
    # The reformatting as C-contiguous below is a work around for a
    # bug in Rasterio 0.10.
    dst.write_band(1, np.asarray(np.rot90(Z16), order='C'))
```

```python
schema = { 'geometry': 'Polygon', 'properties': { 'name': 'str' } }
with fiona.collection("crimepolys-shapely.shp", "w", "ESRI Shapefile", schema, crs=from_epsg(3857)) as outshp:
    for feature, shapes in features.shapes(asarray(rot90(Z16.astype(uint8)),order='C'),transform=transform):
        tfeat = []
        for i in feature['coordinates'][0]:
            tfeat.append((i[0],i[1]))
        tfeat = array(tfeat)
        poly = Polygon(tfeat)
        outshp.write({'geometry': mapping(poly),'properties': {'name': i}});
```
